```python
import matplotlib.pyplot as plt

class_names = [
    'adipose',
    'background',
    'debris',
    'lymphocytes',
    'mucus',
    'smooth muscle',
    'normal colon mucosa',
    'cancer-associated stroma',
    'colorectal adenocarcinoma epithelium'
]

class_counts = [
    9366,
    9509,
    10360,
    10401,
    8006,
    12182,
    7886,
    9401,
    12885
]

plt.figure(figsize=(10, 6))
bars = plt.bar(range(len(class_counts)), class_counts, color='skyblue', edgecolor='black')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 200, str(height), ha='center', va='bottom', fontsize=8)

plt.xticks(range(len(class_counts)), class_names, rotation=45, ha='right')
plt.ylabel('Number of Data Points')
plt.title('Class Distribution (Balance) in Dataset')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
